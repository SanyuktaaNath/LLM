```python
# Calling of libraries
import pandas as pd
import numpy as np
from  pathlib import Path

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Displaying of multiple command output from a cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
data = pd.read_csv('/content/sample_data/california_housing_train.csv')
data.head()
```

```python
# Get target and predictors: # Here we assign 'Median_House_Value' in Output variable
y = data.pop("median_house_value")
X = data
```

```python
model = tf.keras.Sequential()
```

```python
model.add(layers.Input(shape = (8,) ))
```

```python
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(32, activation='selu'))
model.add(layers.Dropout(0.4))
```

```python
model.add(layers.Dense(1,activation='linear'))
```

```python
model.compile( loss = "mae")
```

```python
model.fit(X,y, epochs = 200 )
```
